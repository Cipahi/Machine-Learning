# Generalized Linear Model

We have seen that in linear regression, we are trying to estimate the coefficient of each predictors by solving the least square equation. Also, in logistic regression, we model on the log odds of y by fitting a linear model on it instead of just fit a linear model on y. 

From Generalized Linear Model(GLM), it turns out that both linear regression and logistic regression is just a special case of GLM. The great thing about GLM is that they allow us to use 'response' that can take any value. For example, in logistic regression, the response is 0 or 1; in poisson regression, our response variable is count data.

Then the x's and the y are connected via the so-called "Link function". We use these "link functions" in the instances in which the x's are not related to the y's in a linear manner. For example, the link function of logistic regression is the log-odd (or logit), which has a linear relationship with the x's. 

Another way to explain it is that rather than force a linear relationship between the x's and the y's, the GLM allows the x's related to y's through the link function.

### Poisson Regression

Let's first look at what Poisson distribution is.

**Poisson Distribution**

The formula of Poisson distribution is as follows:
$$f(y;\lambda) = \frac{e^{-\lambda}\lambda^y}{y!}$$

where y = 0,1,2,$\dots$

Let's take a look at how the distribution look like:

```r
ggplot(data.frame(x=c(0:10)), aes(x)) +
    stat_function(geom="line", n=11, fun=dpois, args=list(1),aes(colour = "lambda 1"), size = 1.5) +      
    stat_function(geom="line", n=11, fun=dpois, args=list(2),aes(colour = "lambda 2"), size = 1.5) + 
    stat_function(geom="line", n=11, fun=dpois, args=list(4),aes(colour = "lambda 4"), size = 1.5) +
    stat_function(geom="line", n=11, fun=dpois, args=list(10),aes(colour = "lambda 10"), size = 1.5) +
    scale_colour_brewer(palette="Accent") +
        labs(colour = "Groups")
```

* Poisson distribution only take x values >= 0.
* The occurrence of one event does not affect the probability that a second event will occur. That is, events occur independently
* The rate at which events occur is constant. The rate cannot be higher in some intervals and lower in other intervals.

Let's take the example from wiki to illustrate it:

*For instance, an individual keeping track of the amount of mail they receive each day may notice that they receive an average number of 4 letters per day. If receiving any particular piece of mail does not affect the arrival times of future pieces of mail, i.e., if pieces of mail from a wide range of sources arrive independently of one another, then a reasonable assumption is that the number of pieces of mail received in a day obeys a Poisson distribution.*

From the above plot, we can then say that the probability of receiving 4 letters in a day is 0.2; the probability of receiving 2 letters in a day is 0.15. We can use the probabilty generated from Poisson distribution to be the probability of some event's happening. Another thing to note is that if we want to know the probability, for example, what's the probabilty of receiving 20 letters per week. We'll need to change the lambda into 28, since the average number of letters per week will be 28 given that there is on average 4 letters per day. Then we use the poisson distribution with $\lambda= 28$ by pluging $x=20$ to get our final probability.

Ok, we have known enough about poisson distribution, then what about poisson regression? If our response variable has a poisson distribution, which means it is count data, then we can use poisson regression to model it.
